### Using `value_counts()` method

```python
len(pokemon)    # number of elements in the Series
```

```python
pokemon.value_counts()
```

```python
len(pokemon.value_counts()) # number of unique values in the Series
```

```python
pokemon.value_counts(normalize=True) * 100    # relative frequencies of unique values
```

```python
google.min()
```

```python
google.max()
```

```python
len(google)
```

```python
buckets = [0, 200, 400, 600, 800, 1000, 1200, 1400] # define bins
```

```python
google.value_counts(bins=buckets, sort=False).sort_index() # count of values in each bin
```

```python
google.value_counts(bins=6).sort_index()
```

```python
battles.value_counts(dropna=False)
```

```python
battles.index.value_counts()
```

### Using the `apply` method to apply different function on the Series object

```python
funcs = [len, min, max]
for func_name in funcs:
    print(func_name(google))
```

```python
google
```

```python
google.apply(round)
google.apply(func=round)
```

```python
pokemon
```

```python
def single_or_multi(pokemon_type):
    if "/" in pokemon_type:
        return "Multi"
    else:
        return "Single"

pokemon.apply(single_or_multi).value_counts()
```
