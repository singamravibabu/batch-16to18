# Array Operations and Broadcasting

```python
import numpy as np
```

```python
a1 = np.array([1, 2, 3])
a2 = np.array([4, 5, 6])
```

```python
print(a1 + a2)  # Element-wise addition
print(a1 * a2)  # Element-wise multiplication
print(a1 - a2)  # Element-wise subtraction
print(a1 / a2)  # Element-wise division
```

```python
a1 + 10 # Broadcasting: adding scalar to array
```

```python
arr2 = np.array([1, 4, 9, 16])
print(np.sqrt(arr2))  # Square root of each element
print(np.log(arr2))   # Natural logarithm of each element
print(np.exp(arr2))   # Exponential of each element
print(np.sin(arr2))   # Sine of each element (in radians)
print(np.cos(arr2))   # Cosine of each element (in radians)
```

```python
arr4 = np.array([[10, 20, 30], [40, 50, 60]])
np.sum(arr4)
```

```python
np.sum(arr4, axis=0)  # Sum along columns
```

```python
np.sum(arr4, axis=1)  # Sum along rows
```

```python
np.mean(arr4)  # Mean of all elements
np.max(arr4)   # Maximum element
np.min(arr4)   # Minimum element
np.std(arr4)   # Standard deviation of all elements
```
