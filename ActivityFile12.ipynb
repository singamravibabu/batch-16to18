# GroupBy Object

```python
import numpy as np
import pandas as pd
```

```python
fortune = pd.read_csv("data/fortune1000.csv")
fortune
```

```python
sectors = fortune.groupby("Sector")
sectors
```

```python
sectors.groups
```

```python
sectors.first()
```

```python
sectors.last()
```

```python
sectors.nth(5) # sectors.head() and sectors.tail()
```

```python
sectors.sum()
```

```python
sectors.nth(3)
```

```python
sectors.get_group("Apparel").sum()
```

```python
sectors["Revenues"].sum()
```

```python
sectors_industries = fortune.groupby(["Sector", "Industry"])
```

```python
aggregations = {
    "Profits": "sum",
    "Revenues": "mean",
    "Employees": "count"
}
sectors.agg(aggregations)
```
