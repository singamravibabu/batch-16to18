```python
import numpy as np
import pandas as pd
```

```python
# from datetime import date, time, datetime, timedelta
import datetime as dt
```

```python
# dt.date(year, month, day)
# dt.date.today()   # current date
dt.date(2025, 11, 25)
```

```python
# dt.datetime(year, month, day, hour, minute, second, microsecond)
# dt.datetime.now()  # current date and time
dt.datetime(2025, 11, 25, 10, 30, 45)
```

```python
# dt.time(hour, minute, second, microsecond)
dt.time(14, 45, 30)
```

```python
pd.Timestamp('2005-7/17 3:20 PM')
```

```python
disney["Date"] + pd.DateOffset(years=40, months=3, days=4, hours=10, minutes=30, seconds=15)
```
