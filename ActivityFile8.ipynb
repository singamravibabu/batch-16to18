# FILTERING DATAFRAMES

```python
import numpy as np
import pandas as pd
```

```python
employees = pd.read_csv("data/employees.csv", parse_dates=["Start Date"])
employees.head(3)
```

```python
employees.info()
```

### Using `astype()` method() change the datatype of a column

```python
employees["Mgmt"] = employees["Mgmt"].astype(bool)
employees.info()
```

```python
employees["Salary"] = employees["Salary"].fillna(0).astype(int)
employees.info()
```

```python
employees["Gender"] = employees["Gender"].astype('category')
employees.info()
```

```python
employees["Team"] = employees["Team"].astype('category')
employees.info()
```

```python
employees.shape
```

### Filtering the DataFrame using a Single Condition

```python
marias = employees["First Name"] == "Maria"
```

```python
employees[employees["First Name"] == "Maria"]
```

```python
employees[marias]
```

```python
finance = employees["Team"] == "Finance"
employees[finance]
```

```python
employees[employees["Mgmt"]]
```

```python
employees[~(employees["Mgmt"])]    # ~ Inversion operators
```

```python
greater_than_100k = employees["Salary"]>100000
employees[greater_than_100k]
```

```python
# & for and; | for or
is_female = employees["Gender"] == "Female"
is_biz_dev = employees["Team"] == "Business Dev"
is_manager = employees["Mgmt"]
employees[is_female & is_biz_dev].head(3)
employees[is_female & is_biz_dev & is_manager].head(3)
```

```python
# isin() method
star_teams = ["Legal", "Sales", "Marketing"]
all_star_teams = employees["Team"].isin(star_teams)
employees[all_star_teams]
```

```python
salary_80K_90K = employees["Salary"].between(80000, 90000)
employees[salary_80K_90K]
```

```python
between_80_90 = employees["Start Date"].between(
    left="1980-01-01",
    right="1990-01-01"
)
employees[between_80_90]
```

```python
start_with_r = employees["First Name"].between("R", "S")
employees[start_with_r]
```

```python
employees.info()
```

```python
# isnull() identifies null values in a columns
employees["First Name"].isnull()
employees[employees["First Name"].isnull()]
```

```python
non_null_names = ~(employees["First Name"].isnull())
employees[non_null_names]
```

```python
# notnull() used to find non-missing values (not null values)
team_exists = employees["Team"].notnull()
employees[team_exists]
```

```python
# dropna() - you can drop all the rows with missing values - NaN or NaT
employees.dropna()      # drop all the rows with missing a value
```

```python
employees.dropna(how="any")         # if at least one missing value in a row drop such rows
```

```python
employees.dropna(how="all")     # only when all the values in a row are missing
```

```python
employees.dropna(subset=["Gender"])
```

```python
employees.dropna(subset=["First Name", "Gender"])
```

```python
# thresh specifies number of non-missing values to keep a row from deleting
employees.dropna(thresh=4)
```

#### Handling duplicate rows

```python
# duplicated() method returns False for non-duplicated values and True for duplicates
employees["Team"].duplicated(keep="first")
```

```python
employees["Team"].duplicated(keep="last")
```

```python
employees[~employees["Team"].duplicated()]  # show starting rows for Team column without duplicates
```

```python
employees[~employees["Team"].duplicated(keep="last")] # show last non-duplicates
```

#### drop_duplicates() method to drop duplicate rows

```python
employees.drop_duplicates()     # drop duplicate rows
```

```python
employees.drop_duplicates(subset=["Team"])
```
