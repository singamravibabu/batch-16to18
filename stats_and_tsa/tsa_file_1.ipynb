# Introduction to Descriptive Statistics

## 2. Load the Dataset

````python
import pandas as pd
df = pd.read_csv("sales_daily.csv")
df.head()
````

````python
# Conver the date column:
df['date'] = pd.to_datetime(df['date'], format="mixed")
````

## 3. Central Tendency

````python
# mean (average)
df['sales'].mean()
````

````python
# median
df['sales'].median()
````

````python
# mode
df['sales'].mode()
````

## 4. Measures of Dispersion

````python
# range: difference between largest and smallest value
df['sales'].max() - df['sales'].min()
````

````python
# Variance
df['sales'].var()
````

````python
# Standard Deviation
df['sales'].std()
````

````python
# Interquartile Range(IQR)
Q1 = df['sales'].quantile(0.25)
Q2 = df['sales'].quantile(0.5)
Q3 = df['sales'].quantile(0.75)
# print(f"Q1: {Q1}, Q2: {Q2}, and Q3: {Q3}")
IQR = Q3 - Q1
print("IQR:", IQR)
````

## 5. Detecting Outliers (Using IQR Rule)

````python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['sales'] < lower) | (df['sales'] > upper)]
outliers
````

## 6. Data Visualization (Basic)

````python
import matplotlib.pyplot as plt

df['sales'].hist(bins=20)
plt.title("Histogram of Daily Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
````

````python
# box plot
df.boxplot(column='sales')
plt.title("Boxplot of Daily Sales")
plt.show()
````
