### Special Arrays

```python
arr0 = np.zeros((3, 5), dtype='int')    # An array with 3x5 shape and all values as 0
print(arr0)
```

```python
arr1s = np.ones((2, 8), dtype='float')   # An array with 2x8 shape and all values as 1.0
print(arr1s)
```

```python
arrfull = np.full((4, 4), 7)               # An array with 4x4 shape and all values as 7
print(arrfull)
```

```python
arrId = np.eye(4)
print(arrId)
```

```python
arrSeq = np.arange(5)
arrSeq
```

```python
arrSeq2 = np.arange(2, 10, 2)
arrSeq2
```

```python
equally_spaced_values = np.linspace(1, 3, 20)
equally_spaced_values
```

```python
arr_r = np.random.rand(3, 4)   # 3x4 array with random values between 0 and 1 [0, 1)
arr_r
```

```python
np.random.randn(2, 3)   # 2x3 array with random values from standard normal distribution
```

```python
np.random.randint(5, 10, size=(4, 20))   # 3x7 array with random integers between 5 and 10 [5, 10)
```
