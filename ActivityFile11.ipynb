# Pivot Tables

```python
import numpy as np
import pandas as pd
```

```python
sales = pd.read_csv("data/sales_by_employee.csv", parse_dates=["Date"])
sales
```

```python
sales.pivot_table(
    index="Name",
    columns="Customer",
    values="Revenue",
    aggfunc="sum",
    fill_value=0,
    margins=True,
    margins_name="Total"
)
```
