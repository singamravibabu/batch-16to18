# Pivot Tables

```python
import numpy as np
import pandas as pd
```

```python
sales = pd.read_csv("data/sales_by_employee.csv", parse_dates=["Date"])
sales
```

```python
sales.pivot_table(
    index="Name",
    columns="Customer",
    values="Revenue",
    aggfunc="sum",
    fill_value=0,
    margins=True,
    margins_name="Total"
)
```

```python
p1 = sales.pivot_table(
    index="Name",
    columns="Customer",
    values="Revenue",
    aggfunc="sum",
)
p2 = sales.pivot_table(
    index=["Name", "Date"],
    values="Revenue",
    aggfunc="sum",
)
# stack(), unstack()
```

```python
p1.stack()
```

```python
p2.unstack()
```

### Melting a Dataset

```python
vgs = pd.read_csv("data/video_game_sales.csv")
vgs
```

```python
vgs.melt(id_vars="Name", value_vars="NA")
```

```python
vgs.melt(id_vars="Name", value_vars=["NA", "EU", "JP", "Other"],
         var_name="Region", value_name="Sales")
```
